### PDF Processing Tool

First, let's install the `PyPDF2` library, which is commonly used for reading and manipulating PDF files.

```python
!pip install PyPDF2
```

Now, let's add code to allow you to upload a PDF file. After execution, a button will appear to browse and select your PDF.

```python
from google.colab import files
import PyPDF2
import io

uploaded = files.upload()

for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')
    pdf_file = io.BytesIO(uploaded[filename])

    # Read the PDF content
    reader = PyPDF2.PdfReader(pdf_file)
    pdf_text = ""
    for page in reader.pages:
        pdf_text += page.extract_text() + "\n"

    print("\n--- Extracted PDF Text (first 500 characters) ---")
    print(pdf_text[:200])
    print("\n--- End of preview ---")

    # Store the full text for later use
    # You can pass this `pdf_text` variable to your summarization model
    summarization_input_text = pdf_text
    print(f"PDF content loaded into `summarization_input_text` variable. Total characters: {len(summarization_input_text)}")
```

```python
from pprint import pprint

System_promt="""You are capable agent with proper sound logical and literary skill. You respond to every query deligently without missing key-points.
                Always provide answer in bullet format, and only provide answer (DO NOT ADD THE QUESTION IN FINAL ANSWER ).
                your process include-
                Thinking>action>observation>final answer(this process can repeat n times to attain desired result.)
                Action- This process does not involve external tools.
                """
completion = client.chat.completions.create(
    model="openai/gpt-oss-120b:groq",
    messages=[

           {"role":"system","content":System_promt},
           { "role": "user","content": f"summarize this pdf in paragraph points in 500 words, including the underlying meaning. Here is the content: {summarization_input_text}"},

    ],
    max_tokens=350,
    temperature=0.6,
    tool_choice="none"
)


pprint(completion.choices[0].message.content)
```
