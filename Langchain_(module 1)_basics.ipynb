```python
from google.colab import userdata
#langchain provides wrapper to use its ecosystem
from huggingface_hub.inference._providers import openai
from langchain_openai.chat_models import ChatOpenAI
from huggingface_hub import InferenceClient
import os

# initializing Hugging face client
client= InferenceClient(model="openai/gpt-oss-120b")
#wrap it in Lancchain's chatopenai interface
model=ChatOpenAI(
    openai_api_key=userdata.get('HF_TOKEN'),
    model_name="openai/gpt-oss-120b",
    base_url="https://router.huggingface.co/v1",
    max_tokens=300 # Setting max_tokens here
)
```

**Multi conversation agent**

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
agent= create_agent(model=model)
#HumanMessage=Its langchain's agent abstraction| it internally manages converstion state,indexing, response formatting.
#agent is designed to simulate a chatloop
question=HumanMessage(content="What is the capital of the moon?")
result=agent.invoke(
  {"messages":[question]})

print(result['messages'][1].content)# indexing is [1]as it shows user output rather that user input which is in [0]| in case of multiple back-forth indexing can require 3,4... so on.
```

```python
from pprint import pprint
pprint(result)
```

```python
from langchain.messages import AIMessage, HumanMessage

result=agent.invoke(  #invoke provide entire answer| while .stream() provide chunks of answer
    {'messages':[
        HumanMessage(content='what is the capital of Moon?'),
        AIMessage(content='The capital of the Moon is Luna city.'),
        HumanMessage(content='Interseting, tell me more about Luna')
    ]}
)
pprint(result)
```

**Promting**

```python
from langchain.agents  import create_agent
from langchain.messages  import HumanMessage, SystemMessage# we are importing it,because it comes from Langchain library rather than python.

system_promt="""You are a science fiction writer, create a capital city at the user request."""#(Basic promt)

scifi_agent= create_agent (model=model)

result=scifi_agent.invoke(
    {'messages':[
        SystemMessage(content=system_promt),
        HumanMessage(content='what is the capital of moon?')
    ]}
)

print(result['messages'][-1].content)#[-1] shows last reply of AI
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage, SystemMessage

system_promt="""You are a science fiction writer, create a capital city at the user request.
Please keep to the below structure.
Name: The name of the capital city
Location: Where it is based
Vibe: 5-6 words to describe its vibe
Economy: Main industries """ #(Structured Promt)

scifi_agent= create_agent (model=model)

result=scifi_agent.invoke(
    {'messages':[
        SystemMessage(content=system_promt),
        HumanMessage(content='what is the capital of Venus?')
    ]}
)

print(result['messages'][-1].content)
```

Structured Output -using pydantic | parsing= taking raw string and converting into a structured object that you can use in code.

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage, SystemMessage
from pydantic import BaseModel
from langchain_core.output_parsers import PydanticOutputParser # This parser knows how to take raw text and turn into a Capitalinfo object
#sometime in langchain create_agent dosen't natively accept Output_parser

class CapitalInfo(BaseModel):
  name:str
  location:str
  vibe:str
  economy:str

# Initialize the PydanticOutputParser with your Pydantic model
parser = PydanticOutputParser(pydantic_object=CapitalInfo)

# Integrate the format instructions into your system prompt
system_promt="""You are a science fiction writer, create a capital city at the user request.
{format_instructions}
""".format(format_instructions=parser.get_format_instructions())


scifi_agent= create_agent (model=model)

result=scifi_agent.invoke(
    {'messages':[
        SystemMessage(content=system_promt),
        HumanMessage(content='what is the capital of venus?')
    ]}
)

# Parse the agent's string response into the Pydantic object
# The agent's response is in the last message's content
capital_info = parser.parse(result['messages'][-1].content)

print(f'Name: {capital_info.name}')
print(f'Location: {capital_info.location}')
print(f'Vibe: {capital_info.vibe}')
print(f'Economy: {capital_info.economy}')
```

**Tools**

```python
from langchain.tools import tool

@tool   # Basic form
def square_root(x:float)->float:
  "calculate the square root of a number"
  return x**0.5

square_root.invoke({'x':36})
```

```python
from langchain.tools import tool

@tool('square_root') # explicit form(useful when funtn name dosen't match tool name)
def  tool1(x:float)->float:
  'Calculate the square of any number'
  return x**0.5

tool1.invoke({'x':44})
```

```python
from langchain.tools import tool
@ tool ('square_root', description='calculate the square of any number')#explicit and description , provide full control over tool registration
def tool1(x: float) :
  return x**0.5

square_root.invoke({'x':256})#tool1 can also be used
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage , SystemMessage

system_prompt="""You are a math wizard and can provide solution to any math question.
                 In this case square root. You can calculate by using provided tool"""

math_agent=create_agent(model=model,
                        tools=[square_root])

result=math_agent.invoke(
    {'messages':[
        SystemMessage(content=system_prompt),
        HumanMessage(content='what is the square root of 225?')
    ]}
)
print(result['messages'][-1].content)
```

**Web search tools**


*Search = "Find me something."

*Extract = "Pull out the useful bits."

*Map = "Convert or link data."

*Crawl = "Systematically gather everything."  



```python
from langchain.tools import tool
from typing import Dict, Any
import os
!pip install tavily-python
from tavily import TavilyClient
from google.colab import userdata
os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')
tavily_client=TavilyClient()

@tool ('web_search_tool', description='search the web for information')
def ws_tool(query: str):
  return tavily_client.search(query,max_results=3)#it limits the results

ws_tool.invoke('what are top countries with highest gdp/capita?')
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pprint import pprint
agent=create_agent (model=model,
                  tools=[ws_tool]
                   )

question=HumanMessage(content='give me top 5 countries with highest gdp/capita')

result=agent.invoke(
   { 'messages':[question]}
)
pprint(result['messages'])
```

**Memory**

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from langgraph.checkpoint.memory import InMemorySaver

system_prompt="""You are a helpful assistant. You MUST remember and utilize all previous details from our conversation to answer my questions accurately and contextually. Always refer to past information when relevant."""

agent=create_agent(model=model,
                   system_prompt=system_prompt,
                   checkpointer=InMemorySaver())
question=HumanMessage(content='My name is AK, I like spicy food like shwawarma, burger, fries and in driks I like beer.')
config={'configurable':{'thread_id':'1'}}

result=agent.invoke(
    {'messages':[question]},
    config,
)
print(result['messages'][-1].content)
#pprint(result[message]) can give you details about tokenusage , output-token and other things ,but this provide clean output.
```

```python
question=HumanMessage(content='what is my name and what i like to drink')
result=agent.invoke(
    {'messages':[question]},
    config,
)
print(result['messages'][-1].content)
```
